```python
# Replace the file path with the correct one
file_path = "test.csv"

# Try reading the file with 'latin1' encoding

df1 = pd.read_csv(file_path, encoding='latin1')
# Further processing or modifications as needed
df1.dropna(inplace = True)

print(df1)
```

# Cleaning of the data

```python
# Tokenize
# "I am a python dev" -> ["I", "am", "a", "python", "dev"]
```

```python
from nltk.tokenize import RegexpTokenizer
# NLTK -> Tokenize -> RegexpTokenizer
```

```python
# Stemming
# "Playing" -> "Play"
# "Working" -> "Work"
```

```python
from nltk.stem.porter import PorterStemmer
# NLTK -> Stem -> Porter -> PorterStemmer

from nltk.corpus import stopwords
# NLTK -> Corpus -> stopwords
```

```python
# Downloading the stopwords
import nltk
nltk.download('stopwords')
```

```python
tokenizer = RegexpTokenizer(r"\w+")
en_stopwords = set(stopwords.words('english'))
ps = PorterStemmer()
```

```python
def getCleanedText(text):
  text = str(text)
  text = text.lower()

  # tokenizing
  tokens = tokenizer.tokenize(text)
  new_tokens = [token for token in tokens if token not in en_stopwords]
  stemmed_tokens = [ps.stem(tokens) for tokens in new_tokens]
  clean_text = " ".join(stemmed_tokens)
  return clean_text
```

# Input from the user

```python
x_train = df['text']
y_train = df['sentiment']
x_test = df1['text']
y_test = df1['sentiment']
```

```python
X_clean = [getCleanedText(i) for i in x_train]
xt_clean = [getCleanedText(i) for i in x_test]
```

# Vectorize

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(ngram_range = (1,2))
# "I am PyDev" -> "i am", "am Pydev"
```

```python
X_vec = cv.fit_transform(X_clean).toarray()
```

```python
X_vec
```

```python
Xt_vect = cv.transform(xt_clean).toarray()
```

```python
Xt_vect
```

# Multinomial Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
mn = MultinomialNB()
```

```python
mn.fit(X_vec, y_train)
```

```python
y_pred = mn.predict(Xt_vect)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')
```
